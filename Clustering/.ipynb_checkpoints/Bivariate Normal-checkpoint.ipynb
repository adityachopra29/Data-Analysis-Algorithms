```python
# prompt: I want to visualize bivariate normal distribution. Give the code, the covariance matrix should be in my control

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal

# Input the covariance matrix
cov_matrix = [[1, -0.5], [-0.5, 1]]  # Example covariance matrix

# Create a grid of points
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Create a multivariate normal distribution object
rv = multivariate_normal([0, 0], cov_matrix)

# Plot the bivariate normal distribution
fig, ax = plt.subplots()
ax.contourf(x, y, rv.pdf(pos))
plt.title('Bivariate Normal Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# prompt: now, I want to bisualize it in 3-D

from mpl_toolkits.mplot3d import Axes3D

# Create a figure and an axes object for the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, rv.pdf(pos), cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Probability Density')
plt.title('3D Bivariate Normal Distribution')

# Display the plot
plt.show()
```

```python
# prompt: plot 3-4 different bivariate normal distributions having different covariance matrices in 3-D

# Define different covariance matrices
cov_matrices = [
    [[1, 0], [0, 1]],
    [[1, 0.5], [0.5, 1]],
    [[1, -0.5], [-0.5, 1]],
    [[2, 0], [0, 1]]
]

# Create a figure and an axes object for the 3D plot
fig = plt.figure(figsize=(10, 8))

# Plot each distribution
for i, cov_matrix in enumerate(cov_matrices):
    ax = fig.add_subplot(2, 2, i + 1, projection='3d')
    rv = multivariate_normal([0, 0], cov_matrix)
    ax.plot_surface(x, y, rv.pdf(pos), cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Probability Density')
    ax.set_title(f'Covariance Matrix {i+1}')

plt.tight_layout()
plt.show()
```

```python
# prompt: generate random dataset having 3 classes red, blue and green and 2 features. The class should be "red, green, and blue

from sklearn.datasets import make_blobs
# Generate a dataset with 3 classes and 2 features
X, y = make_blobs(n_samples=300, centers=3, n_features=2, random_state=0)

# Assign colors to each class
colors = ['red', 'green', 'blue']
color_map = [colors[label] for label in y]

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=color_map)
plt.title('Generated Dataset with 3 Classes')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# prompt: run a logistic regression model on X,y

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, color_map, test_size=0.2, random_state=42)

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)
```
