Optimal Testing Problem

```python
p=0.01
expectedtests = {}
for N in range(2,500):
  expectedtests[N]= (1000/N + 1000*(1-((1-p)**N)))
```

```python
print(expectedtests)
```

```python
import matplotlib.pyplot as plt
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(15)
plt.plot(list(expectedtests.keys())[0:80], list(expectedtests.values())[0:80], 'o')
```
