#To demonstrate the expected value and variance formula that we derived in class for a random variable which is given by a sum of "random number of random variables"

#N = Binomial random variable, hence E(N) = 100, Var(N) = 20

#P = 0, with 0.1 prob P = 1, with 0.8 prob P = 2, with 0.1 prob

#Thus, E(P) = 1300, Var(P) = (5000^2)0.2 + (1000^2)0.3 - 1300^2 = 3610000

#Thus, E(P) = 1, Var(P) = ((-1)^2)0.1 + (0^2)0.8 + (1^2)0.1 = 0.2


#Thus, as per the derived expression, E(R) = 50(3610000)+50((1300)(1300)) = 265000000

#We will now try to verify the results from simulation

```python
Cups = []
for i in range(100000):
  N_cust = np.random.poisson(500)
  totalpurchase = 0
  for j in range(N_cust):
    rpurch = np.random.uniform(0,1)
    if rpurch<0.1:
      totalpurchase = totalpurchase + 0
    elif rpurch<0.9:
      totalpurchase = totalpurchase + 1
    else:
      totalpurchase = totalpurchase + 2
  Cups.append(totalpurchase)  
```

```python
print(Cups)
```

```python
average = sum(Cups)/len(Cups)
average
```

```python
np.var(Cups)
```

```python
var = sum((x-average)**2 for x in Cups) / len(Cups)
var
```

```python
varest = var*(len(Cups)/(len(Cups)-1))
varest
```

```python
#Poisson Distribution
import math
```

```python
poissonprob = {}
lamdba = 50
for i in range(100):
    poissonprob[i] = np.exp(-lamdba)*(lamdba**i)/(math.factorial(i))
```

```python
np.random.poisson(lamdba)
```

```python
poissonprob
```

```python
import matplotlib.pyplot as plt
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(15)
plt.plot(list(poissonprob.keys())[0:80], list(poissonprob.values())[0:80], 'o')
```
