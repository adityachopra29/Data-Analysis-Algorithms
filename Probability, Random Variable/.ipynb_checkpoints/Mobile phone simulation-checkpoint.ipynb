```python
import random
import pandas as pd

# Read the Excel file
df = pd.read_excel('IBM-322_2024-25_Autumn (4).xlsx')

# Assuming the column with names is called 'Name'
names = df['StudentName'].tolist()
shuffled_names = names.copy()
random.shuffle(shuffled_names)

#print("Original names:", names)
#print("Shuffled names:", shuffled_names)

```

```python
from IPython.display import HTML, display

def print_same_entries(list1, list2):
  for i in range(len(list1)):
    if list1[i] == list2[i]:
      display(HTML(f"<h2>{list1[i]}</h2>"))

print_same_entries(names, shuffled_names)
```

```python
nmatcheslist = []
for i in range(10000):
    shuffled_names = names.copy()
    random.shuffle(shuffled_names)
    nmatches = 0
    for i in range(len(names)):
        if names[i] == shuffled_names[i]:
            nmatches = nmatches + 1
    nmatcheslist.append(nmatches)
      
    
    
```

```python
print(nmatcheslist)
```

```python
import numpy as np
np.average(nmatcheslist)
```

```python
np.var(nmatcheslist)
```
