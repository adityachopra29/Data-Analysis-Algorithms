To obtain the values(CDF, PDF) from a normal distributions

```python
mean = 100
sigma = 2
```

```python
k= 2
norm.cdf(mean+k*sigma,mean,sigma)-norm.cdf(mean-k*sigma,mean,sigma)
```

```python
norm.pdf(mean+k*sigma,mean,sigma)
```
